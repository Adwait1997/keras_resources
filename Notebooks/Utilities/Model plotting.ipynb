# Model plotting utilities

# plot_model

syntax: tf.keras.utils.plot_model(model, to_file="model.png", show_shapes=False, show_dtype=False, show_layer_names=True, rankdir="TB", expand_nested=False, dpi=96)
    
1. Converts a Keras model to dot format and save to a file.   

# Arguments

1. model: A keras model
    
2. to_file: file name 
    
3. show_shapes: whether to display shape info
    
4. show_dtype: whether to display layer dtypes
    
5. show_layer_names: whether to display layer names
    
6. rankdir: rankdir argument passed to PyDot, a string specifying the format of the plot: 'TB' creates a vertical plot; 'LR' creates a horizontal plot.
        
7. expand_nested: expand nested models into cluster
    
8. dpi: dots per inch
    
# Returns

A Jupyter notebook Image object if Jupyter is installed. This enables in-line display of the model plots in notebooks.

# model_to_dot function

syntax: tf.keras.utils.model_to_dot(model,show_shapes=False,show_dtype=False,show_layer_names=True,rankdir="TB",expand_nested=False,dpi=96,subgraph=False)
    
1. Convert a Keras model to dot format.


# Arguments

1. subgraph: whether to return a pydot.Cluster instance.


# Returns

1. A pydot.Dot instance representing the Keras model or a pydot.Cluster instance representing nested model if subgraph=True.

```python
# Examples
import tensorflow as tf

input = tf.keras.Input(shape=(100,), dtype='int32', name='input')
x = tf.keras.layers.Embedding(
    output_dim=512, input_dim=10000, input_length=100)(input)
x = tf.keras.layers.LSTM(32)(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
output = tf.keras.layers.Dense(1, activation='sigmoid', name='output')(x)
model = tf.keras.Model(inputs=[input], outputs=[output])
dot_img_file = '/tmp/model_1.png'
tf.keras.utils.plot_model(model, to_file=dot_img_file, show_shapes=True)
```
